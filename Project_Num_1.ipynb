#Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
import pickle
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
Url = "https://raw.githubusercontent.com/kunaldhabhai25-ctrl/DSML_Projects/main/Used_Bikes.csv"
df = pd.read_csv(Url)
print("                                 DATA                                 ")
print("______________________________________________________________________")
print(df)
print("----------------------------------------------------------------------")
print('\n')
print("                                 -INFO-                                 ")
print("______________________________________________________________________")
print(df.info())
print("----------------------------------------------------------------------")
print("                                 -DESCRIBE-                             ")
print("______________________________________________________________________")
print(df.describe())
print("----------------------------------------------------------------------")
print("                                 -BRANDS'-                          ")
print("______________________________________________________________________")
print(df['brand'].value_counts())
print("----------------------------------------------------------------------")
print("                                 -OWNERS'-                          ")
print("______________________________________________________________________")
print(df['owner'].value_counts())
print("----------------------------------------------------------------------")
print("                              -CHECKING NULLS'-                          ")
print("______________________________________________________________________")
print(df.isnull().sum().sum())
print("----------------------------------------------------------------------")
print("                            -BRAND VISUALISATION-                          ")
print("______________________________________________________________________")
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(x=df['brand'],y=df['price'])
plt.xticks(rotation=20)
plt.show()
print('\n')
print("----------------------------------------------------------------------")
print("                            -PRICE VISUALISATION-                          ")
print("______________________________________________________________________")
sns.catplot(x='owner',y='price',kind='bar',data=df)
plt.xticks(rotation=20)
plt.show()
```
